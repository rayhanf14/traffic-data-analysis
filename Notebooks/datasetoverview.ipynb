```python
import pandas as pd
```

```python
df_traffic_data = pd.read_csv('../Dataset/Banglore_traffic_Dataset.csv')
```

```python
df_traffic_data['Date'] = pd.to_datetime(df_traffic_data['Date'])
```

```python
df_traffic_data = df_traffic_data[df_traffic_data['Date'].dt.year >= 2023]
df_traffic_data.reset_index(drop=True, inplace=True)

```
